## Predict Play_Duration 

### 1 Data Preparation

#### 1.1 Features Selection

```python
data=df[['play_time','content_type','platform','category_name','play_duration']]
```

```python
data.isna().sum()
```

#### 1.2 Drop rows where play_duration is NaN

```python
data=data[~data.play_duration.isna()]
data.isna().sum()
```

### 2. Data Preprocessing

#### 2.1 Encode DataFrame

```python
data=pd.get_dummies(data)
data.head()
```

#### 2.2 Resample play_time every  6 hours and sum play_duration and others columns too 

```python
data.set_index('play_time',inplace=True)
data=data.resample('6H').sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data.head()
```

```python
data.play_duration.plot()
```

#### 2.3 Clasifiy play_time in to ['Midnight','Morning', 'Afternoon', 'Night' ]

```python
day=['Midnight','Morning', 'Afternoon', 'Night' ]
for i,j in zip ([0,6,12,18],['Midnight','Morning', 'Afternoon', 'Night' ]):
    data[j]=(data.index.hour==i).astype(float);
data.head()
```

#### 2.4 Add Colomn Days

```python
day=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
for i in range(7):
    data[day[i]]=(data.index.dayofweek==i).astype(float);
data.head()
```

### 3 Data Splitting

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
X=data.drop(columns='play_duration')
y=data.play_duration
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

### 4. Training

#### 4.1 Using Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler,StandardScaler,PowerTransformer
```

```python
params={'algo__fit_intercept': [True, False]}
num_pipe=Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler',MinMaxScaler())
])
cat_pipe=Pipeline([
    ('imputer',SimpleImputer(strategy='most_frequent'))
])
prepo=ColumnTransformer([
    ('numeric',num_pipe,X_train.columns[:-11]),
    ('categoric',cat_pipe,X_train.columns[-11:])
])
pipeline=Pipeline([
    ('prepo',prepo),
    ('algo',LinearRegression())
])
model=GridSearchCV(pipeline,params,cv=5,verbose=1,n_jobs=-1)
model.fit(X_train,y_train)
model.score(X_train,y_train), model.score(X_test,y_test)
```

####  4.2 Model Evaluation Using Residual Plot

```python
from jcopml.plot import plot_residual
```

```python
plot_residual(X_train, y_train, X_test, y_test, model);
```
